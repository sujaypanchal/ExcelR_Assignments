**1. Create a list of 5 integers. Perform and print the result of the following operations:**
append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
nums = [10, 20, 30, 40, 50]
print("Original list:", nums)

nums.append(60)
print("After append:", nums)

nums.extend([70, 80])
print("After extend:", nums)

nums.insert(2, 25)
print("After insert at index 2:", nums)

nums.remove(25)
print("After remove 25:", nums)

popped = nums.pop()
print("Popped element:", popped , "Remaining list:", nums)

nums_copy = nums.copy()
nums_copy.clear()
print("Cleared list:", nums_copy)

print("Index of 30:", nums.index(30))
print("Count of 20:", nums.count(20))

nums.sort()
print("Sorted list:", nums)

nums.reverse()
print("Reversed list:", nums)
```

**6. Given a 'marks' variable, print the grade:**

- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'

```python
marks = 42
if marks >= 90:
    grade = 'A'
elif 75 <= marks <= 89:
    grade = 'B'
elif 50 <= marks <= 74:
    grade = 'C'
else:
    grade = 'Fail'

print("Marks:", marks, "Grade:", grade)
```

**7. Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.**

```python
num = 12
if num > 0:
    if num % 2 == 0:
        print(num, "is positive and even")
    else:
        print(num, "is positive but odd")
elif num == 0:
    print("Number is zero")
else:
    print(num, "is negative")

```

# **Section C: Exploring Pandas**

**11. Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:**

- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows

```python
data = {
    'Name': ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Helen", "Ivy", "Jack"],
    'Age': [23, 25, 30, 22, 29, 24, 31, 28, 26, 27],
    'City': ["NY", "LA", "SF", "NY", "LA", "SF", "NY", "LA", "SF", "NY"],
    'Salary': [50000, 60000, 55000, 62000, None, 58000, 60000, 59000, 60000, 61000]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

print("\nInfo:")
print(df.info())

print("\nDescription:")
print(df.describe())

print("\nSelecting Name and City:")
print(df[['Name','City']])

print("\nDropping City column:")
print(df.drop(columns=['City']))

print("\nFilling missing Salary with mean:")
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
print(df)

print("\nRemoving duplicate rows:")
df.drop_duplicates(inplace=True)
print(df)
```
