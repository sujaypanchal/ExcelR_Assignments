```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading dataset
df_sales = pd.read_csv(r"c:\Users\Admin\Documents\EXCELR\DS Assignments\@\sales_data_with_discounts.csv")
print("First 5 rows of dataset:")
print(df_sales.head())
```

```python
# Identify numerical columns
num_cols = df_sales.select_dtypes(include=np.number).columns.tolist()
print("\nNumerical columns:", num_cols)
```

```python
#Calculating the mean, median, mode, and standard deviation for these columns
for col in num_cols:
    print(f"\n--- {col} ---")
    print("Mean:", df_sales[col].mean())
    print("Median:", df_sales[col].median())
    print("Mode:", df_sales[col].mode().values)
    print("Standard Deviation:", df_sales[col].std())
```

**Visulaization**

```python
# Histograms for each numerical column


# Histogram for Volume
plt.hist(df_sales['Volume'].dropna(), bins=20, edgecolor='black', color='orange')
plt.title("Histogram of Volume")
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.show()
# Inference: Right-skewed, most sales have low volume.


# Histogram for Avg Price
plt.hist(df_sales['Avg Price'].dropna(), bins=20, edgecolor='black', color='orange')
plt.title("Histogram of Avg Price")
plt.xlabel("Avg Price")
plt.ylabel("Frequency")
plt.show()
# Inference: Strongly skewed, few very high prices act as outliers.


# Histogram for Total Sales Value
plt.hist(df_sales['Total Sales Value'].dropna(), bins=20, edgecolor='black', color='orange')
plt.title("Histogram of Total Sales Value")
plt.xlabel("Total Sales Value")
plt.ylabel("Frequency")
plt.show()
# Inference: Right-skewed, most sales values are low with some extreme highs.


# Histogram for Discount Rate (%)
plt.hist(df_sales['Discount Rate (%)'].dropna(), bins=20, edgecolor='black', color='orange')
plt.title("Histogram of Discount Rate (%)")
plt.xlabel("Discount Rate (%)")
plt.ylabel("Frequency")
plt.show()
# Inference: More balanced, many discounts between 1519%.


# Histogram for Discount Amount
plt.hist(df_sales['Discount Amount'].dropna(), bins=20, edgecolor='black', color='orange')
plt.title("Histogram of Discount Amount")
plt.xlabel("Discount Amount")
plt.ylabel("Frequency")
plt.show()
# Inference: Right-skewed, small discounts are most common with few very large discounts.


# Histogram for Net Sales Value
plt.hist(df_sales['Net Sales Value'].dropna(), bins=20, edgecolor='black', color='orange')
plt.title("Histogram of Net Sales Value")
plt.xlabel("Net Sales Value")
plt.ylabel("Frequency")
plt.show()
# Inference: Right-skewed, pattern similar to Total Sales Value with some very large outliers.
```

```python
# Box Plots for each numerical column

# Boxplot for Volume
plt.boxplot(df_sales['Volume'].dropna())
plt.title("Boxplot of Volume")
plt.ylabel("Volume")
plt.show()
# Finding: Most values are concentrated at the lower end, with a few extreme outliers (high volumes).

# Boxplot for Avg Price
plt.boxplot(df_sales['Avg Price'].dropna())
plt.title("Boxplot of Avg Price")
plt.ylabel("Avg Price")
plt.show()
# Finding: Very high outliers present. Majority of data is clustered in a much smaller range.

# Boxplot for Total Sales Value
plt.boxplot(df_sales['Total Sales Value'].dropna())
plt.title("Boxplot of Total Sales Value")
plt.ylabel("Total Sales Value")
plt.show()
# Finding: Highly skewed, long upper whisker with significant outliers at high values.

# Boxplot for Discount Rate (%)
plt.boxplot(df_sales['Discount Rate (%)'].dropna())
plt.title("Boxplot of Discount Rate (%)")
plt.ylabel("Discount Rate (%)")
plt.show()
# Finding: Distribution is more compact, but some variation exists between 5% and 20%. Fewer extreme outliers.

# Boxplot for Discount Amount
plt.boxplot(df_sales['Discount Amount'].dropna())
plt.title("Boxplot of Discount Amount")
plt.ylabel("Discount Amount")
plt.show()
# Finding: Right-skewed, small amounts dominate but large outliers are visible.

# Boxplot for Net Sales Value
plt.boxplot(df_sales['Net Sales Value'].dropna())
plt.title("Boxplot of Net Sales Value")
plt.ylabel("Net Sales Value")
plt.show()
# Finding: Similar to Total Sales Value, highly skewed with many extreme outliers.
```

```python
# Bar Chart Analysis for Categorical Columns

# Identify categorical columns
cat_cols = df_sales.select_dtypes(include='object').columns.tolist()
print("Categorical columns:", cat_cols)

# Bar chart for first categorical column
if len(cat_cols) > 0:
    df_sales[cat_cols[0]].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Bar Chart of {cat_cols[0]}")
    plt.xlabel(cat_cols[0])
    plt.ylabel("Count")
    plt.show()
    print(f"Insights for {cat_cols[0]}:")
    print("- Identify which categories appear most frequently.")
    print("- Check if some categories dominate the dataset.")
    print("- Note any underrepresented categories.\n")

# Bar chart for second categorical column
if len(cat_cols) > 1:
    df_sales[cat_cols[1]].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Bar Chart of {cat_cols[1]}")
    plt.xlabel(cat_cols[1])
    plt.ylabel("Count")
    plt.show()
    print(f"Insights for {cat_cols[1]}:")
    print("- Observe whether distribution across categories is balanced.")
    print("- Detect if one or two categories dominate sales/records.\n")

# Bar chart for third categorical column
if len(cat_cols) > 2:
    df_sales[cat_cols[2]].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Bar Chart of {cat_cols[2]}")
    plt.xlabel(cat_cols[2])
    plt.ylabel("Count")
    plt.show()
    print(f"Insights for {cat_cols[2]}:")
    print("- Find which categories are top contributors.")
    print("- Highlight least frequent categories.\n")

# Bar chart for fourth categorical column (if exists)
if len(cat_cols) > 3:
    df_sales[cat_cols[3]].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Bar Chart of {cat_cols[3]}")
    plt.xlabel(cat_cols[3])
    plt.ylabel("Count")
    plt.show()
    print(f"Insights for {cat_cols[3]}:")
    print("- Look for imbalance across categories.")
    print("- Check whether all categories are evenly distributed.\n")
```

## Conclusion

Our analysis of the `sales_data_with_discounts.csv` dataset gives us a clearer picture of sales and discount patterns.

Most of the numerical variables, such as **Volume**, **Total Sales Value**, **Net Sales Value**, and **Discount Amount**, show that the majority of transactions happen at lower levels, while a few very large transactions stand out as outliers. The **Average Price** column, in particular, reveals some products that are priced much higher than the rest, which skews the overall distribution. On the other hand, the **Discount Rate (%)** looks more stable, with many discounts clustering around 1519%, showing that discounting is applied quite consistently.

The boxplots confirmed the presence of these outliers, especially in sales values and prices, which may represent rare but significant high-value deals or possible data entry errors.

Looking at the categorical variables, the bar charts show that certain categorieswhether regions, products, or sales staffdominate the dataset, while others appear much less frequently. This imbalance suggests that sales are concentrated in specific areas, leaving room to explore and strengthen the weaker categories.

### Key Takeaways
Overall, the dataset tells us that:
- Sales values and volumes are **skewed**, with a few big transactions pulling the numbers upward.  
- **Outliers matter** here and should be carefully handled before moving into predictive modeling.  
- **Discount practices are steady**, mostly falling within a common range.  
- Some categories are **overrepresented**, which highlights both strong performers and areas where there may be growth opportunities.  

In short, the dataset is rich in information but needs preprocessingespecially handling skewed data and outliersbefore it can be used for deeper analysis or building predictive models.
