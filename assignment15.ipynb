## Data preprocessing

### Subtask:
Handle missing values, remove outliers, and scale the features.


**Reasoning**:
I need to handle missing values, remove outliers, and scale the features. I will start by checking for missing values.



```python
print("Missing values before handling:")
print(df.isnull().sum())
```

**Reasoning**:
Since there are no missing values, I will proceed to identify and handle outliers. I will use the IQR method for outlier detection and capping.



```python
import numpy as np

numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
numerical_cols.remove('ID#')
numerical_cols.remove('Award?')

print("Identifying and handling outliers using IQR...")

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Capping outliers
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

print("Outliers handled.")
display(df.describe())
```

**Reasoning**:
Now that missing values and outliers have been handled, I will select the relevant features and scale them using StandardScaler.



```python
from sklearn.preprocessing import StandardScaler

features = [col for col in numerical_cols if col not in ['ID#', 'Award?']]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])

scaled_df = pd.DataFrame(scaled_features, columns=features)

display(scaled_df.head())
```

## Implement dbscan clustering

### Subtask:
Apply DBSCAN to the preprocessed data and experiment with different parameter settings (epsilon, minPts).


**Reasoning**:
Apply DBSCAN to the scaled data and experiment with different parameter settings. Then, select a combination of parameters that yields a reasonable number of clusters and add the cluster labels to the original dataframe.



```python
from sklearn.cluster import DBSCAN

# Experiment with different eps and min_samples
eps_values = [0.3, 0.5, 0.7, 0.9]
min_samples_values = [3, 5, 7, 10]

best_eps = None
best_min_samples = None
best_n_clusters = -1

print("Experimenting with DBSCAN parameters:")
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        dbscan.fit(scaled_df)
        n_clusters = len(set(dbscan.labels_)) - (1 if -1 in dbscan.labels_ else 0)
        n_noise = list(dbscan.labels_).count(-1)

        print(f"eps={eps}, min_samples={min_samples}: {n_clusters} clusters, {n_noise} noise points")

        # Select a parameter combination that yields a reasonable number of clusters
        if n_clusters > 1 and n_clusters < len(scaled_df) * 0.5 and n_noise < len(scaled_df) * 0.5:
             if n_clusters > best_n_clusters:
                best_n_clusters = n_clusters
                best_eps = eps
                best_min_samples = min_samples


# Apply DBSCAN with the chosen parameters
if best_eps is not None and best_min_samples is not None:
    print(f"\nApplying DBSCAN with chosen parameters: eps={best_eps}, min_samples={best_min_samples}")
    dbscan_optimal = DBSCAN(eps=best_eps, min_samples=best_min_samples)
    dbscan_optimal.fit(scaled_df)

    # Add the cluster labels to the original DataFrame
    df['DBSCAN_Cluster'] = dbscan_optimal.labels_

    # Display the head of the DataFrame with the new DBSCAN cluster labels
    display(df.head())
else:
    print("\nCould not find a suitable parameter combination for DBSCAN.")
```
